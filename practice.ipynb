```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cityhousingprices = pd.read_csv('housingpricesbycity.csv')
```

```python
cityhousingprices = cityhousingprices.iloc[[0,23, 14]]
cityhousingprices= cityhousingprices.drop("SizeRank", axis = 1)
cityhousingprices= cityhousingprices.drop("RegionType", axis = 1)
cityhousingprices= cityhousingprices.drop("RegionID", axis = 1)
cityhousingprices= cityhousingprices.drop("StateName", axis = 1)
cityhousingprices = cityhousingprices.T
cityhousingprices.columns = cityhousingprices.iloc[0]
cityhousingprices = cityhousingprices[1:]
cityhousingprices = cityhousingprices.reset_index()
cityhousingprices = cityhousingprices.rename(columns={"index": "Date"})
cityhousingprices.rename_axis(None, axis=1, inplace=True)
```

```python
cityhousingprices['Date'] = pd.to_datetime(cityhousingprices['Date'])
```

```python
cityhousingprices
```

```python
plt.figure(figsize=(10, 6))
plt.plot(cityhousingprices['Date'], cityhousingprices['United States'], color='b', linestyle='-', label='Price')


plt.xlabel('Date')
plt.ylabel('Average House Price ($)')
plt.title('Change in Average House Price Over Time in America')
plt.xticks(rotation=45) 
plt.grid(True)


plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(cityhousingprices['Date'], cityhousingprices['Charlotte, NC'], color='b', linestyle='-', label='Price')


plt.xlabel('Date')
plt.ylabel('Average House Price ($)')
plt.title('Change in Average House Price Over Time in Charlotte')
plt.xticks(rotation=45) 
plt.grid(True)


plt.legend()
plt.tight_layout()
plt.show()
```
